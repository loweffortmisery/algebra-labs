```python
n = 1102158078129785185997827239806
G = IntegerModRing(n).unit_group()
order = G.order()
print(f"{'' if G.is_cyclic() else 'не '}є циклічною")
print(f"порядок = {order}")
gen = G.gens()[0]
gen_ = gen
a = 0

for i in range(2, n):
    if (gcd(i, n) != 1):
        continue
    gen_ = gen
    t = 1
    while(gen_.value() != i):
        gen_ *= gen
        t += 1
    if (gcd(t, order) == 1):
        break
min_gen = gen_    
_min = min_gen.value()
print(f"min_gen = {_min}")
print(f"порядок = {min_gen.order()}")


g_1 = G.random_element()
g_2 = G.random_element()
g_3 = G.random_element()
for g in [g_1, g_2, g_3]:
    k = discrete_log(g, min_gen)

    print(f"\n\n{g.value()=}\n{k=}\n{min_gen.value()}^{k} = {g.value()}")
```
